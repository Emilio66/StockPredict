```python
'''
Fetching transaction data from Tushare
All components' data
'''

import tushare as ts
import pandas as pd
import os
from time import time

def fetch(stock, period='D', year = '2002', start = '2002-01-04', end = '2017-11-30'):
    print ("-------- fetching --- ", year)
    t0 = time()
    for s in stock:
        print ("fetching.. ", s)
        s = str(s)
        data = ts.get_k_data(s, start = start, end = end, ktype='D')
        # calculate change amount
        data['chg_amt'] = 0.
        data['chg_amt'][0] = 0.
        for i in range(1, len(data)):
            data['chg_amt'].iloc[i] = (data['close'].iloc[i] - data['close'].iloc[i-1]) / data['close'].iloc[i-1]
        
        # drop useless data
#         data = data.drop(['open', 'close', 'high', 'low', 'volume', 'code'], axis = 1)
        data = data.drop(['code'], axis = 1)
        #print(data)
        #temporary file
        filename = s+'-'+year+'.csv'
        data.to_csv(filename)
#         df = pd.read_csv(filename, usecols=[1,2], index_col=0, parse_dates=True)
        df = pd.read_csv(filename,  usecols=[1,2,3,4,5,6,7],index_col=0, parse_dates=True)
        # 1st row is just for calculation purpose, not useful
        df = df.drop(df.index[0])
        #print(df)
        # final data file
        df.to_csv('../components/'+year+'/'+s+'.csv') 
        #os.remove(filename)
    t1 = time()
    print("------ DONE --- %s: %.2g sec" % (year, t1 - t0))

indexes = [
    '000001',
    '000905',
    '399001'
]
fetch(indexes, year = '2002',  start = '2001-12-31')

## 2002-01-04 - 2017-11-30 ##
stock02 = [['600028',
'600011',
'600019',
'600000',
'600377',
'600016',
'600642',
'600663',
'600188',
'600098'
],[8.74,
2.05,
1.77,
1.37,
1.25,
1.18,
0.78,
0.71,
0.7,
0.7
]]

stock03 = [['600028',
'600050',
'600011',
'600019',
'600377',
'600036',
'600000',
'600016',
'600663',
'600642'
],[8.3,
2.09,
2.04,
2.03,
1.85,
1.83,
1.41,
0.94,
0.71,
0.7
]]
stock04 = [['600028',
600019,
600011,
600050,
600900,
600036,
600377,
600000,
600104,
600016
],[11.61,
2.97,
2.65,
2.6,
2.29,
2.07,
1.4,
1.38,
1.16,
1.08,
]]
stock05 = [[600028,
600019,
600900,
600011,
600050,
600036,
600005,
600009,
600016,
600018,
],[11.72,
2.89,
2.65,
2.51,
2.49,
2.2,
1.22,
1.13,
1.08,
1.06
]]
stock06 = [[600028,
600019,
600036,
600050,
600900,
600011,
600000,
600016,
600009,
600377
],[14.11,
3.12,
2.94,
2.57,
2.45,
2.24,
1.65,
1.28,
1.2,
1.05
]]
stock07 = [[601398,
601988,
600028,
600036,
600019,
600018,
601006,
600016,
600050,
600000
],[21.73,
13.48,
8.9,
2.75,
2.12,
2.05,
1.47,
1.45,
1.38,
1.3
]]
stock08 = [[601857,
601398,
600028,
601628,
601988,
601088,
601318,
600036,
601328,
601601
],[18.58,
7.56,
6.07,
4.47,
4.36,
4.01,
1.88,
1.77,
1.5,
1.41
]]
stock09 = [[601857,
601398,
601988,
600028,
601628,
601088,
600036,
600900,
601318,
601328
],[16.93,
9.14,
5.43,
5.05,
3.99,
2.97,
1.51,
1.39,
1.31,
1.26
]]
stock10 = [[601857,
601398,
600028,
601988,
601628,
601088,
600036,
601318,
601328,
601998
],[12.12,
7.39,
5.34,
4.17,
3.57,
3.11,
1.53,
1.43,
1.31,
1.19
]]
stock11 = [[601857,
601398,
601288,
601988,
600028,
601628,
601088,
601318,
600036,
600000 ,   
],[10.15,
6.21,
4.4,
3.53,
3.15,
2.48,
2.28,
1.5,
1.26,
0.99,
]]

stock12 = [[601857,
601398,
601288,
601988,
600028,
601628,
601088,
600036,
600519,
601318,
],[10.63,
7.49,
5.19,
3.85,
3.38,
2.82,
2.48,
1.41,
1.35,
1.11,
]]
stock13 = [[601857,
601398,
601288,
601988,
600028,
601628,
601088,
600036,
600519,
601318,
],[9.22,
6.87,
5.19,
3.6,
3.05,
2.81,
2.63,
1.53,
1.37,
1.37
]]
stock14 = [[601857,
601398,
601288,
601988,
600028,
601628,
601088,
600036,
601318,
601166,
],
           [8.26,
6.25,
4.82,
3.39,
2.7,
2.08,
1.73,
1.49,
1.32,
1.28,
]]
stock15 = [[601857,
601398,
601288,
601988,
601628,
600028,
601318,
600036,
601088,
600030
],[7.17,
5.28,
4.47,
3.33,
2.91,
2.43,
1.47,
1.4,
1.37,
1.37
]]
stock16 = [[601857,
601398,
601288,
601988,
601628,
600028,
601318,
600036,
600000,
601166,
],
           [4.58,
4.18,
3.22,
2.86,
2,
1.61,
1.32,
1.26,
1.15,
1.1
]]
stock17 = [[601857,
601398,
601288,
601988,
600028,
601628,
600519,
601318,
600036,
600000,
],
          [4.52,
4.18,
3.2,
2.55,
1.82,
1.76,
1.47,
1.35,
1.28,
1.23
]] 
# fetch(stock02, year = '2002', start = '2001-12-31', end = '2002-12-31')
# fetch(stock03, year = '2003', start = '2002-12-31', end = '2003-12-31')
# fetch(stock04, year = '2004', start = '2003-12-31', end = '2004-12-31')
# fetch(stock05, year = '2005', start = '2004-12-31', end = '2005-12-30')
# fetch(stock06, year = '2006', start = '2005-12-30', end = '2006-12-29')
# fetch(stock07, year = '2007', start = '2006-12-29', end = '2007-12-28')
#fetch(stock08[0], year = '2008', start = '2007-12-28', end = '2008-12-31')
# fetch(stock09, year = '2009', start = '2008-12-31', end = '2009-12-31')
# fetch(stock10, year = '2010', start = '2009-12-31', end = '2010-12-31')
# fetch(stock11, year = '2011', start = '2010-12-31', end = '2011-12-30')
# fetch(stock12, year = '2012', start = '2011-12-30', end = '2012-12-31')
# fetch(stock13, year = '2013', start = '2012-12-31', end = '2013-12-31')
# fetch(stock14, year = '2014', start = '2013-12-31', end = '2014-12-31')
# fetch(stock15, year = '2015', start = '2014-12-31', end = '2015-12-31')
# fetch(stock16, year = '2016', start = '2015-12-31', end = '2016-12-30')
# fetch(stock17, year = '2017', start = '2016-12-30', end = '2017-11-30')
```

```python
'''
Locate the weight for each component stock
'''
def find_weight(stocks, s):
    s = str(s)
    for i in range(len(stocks[0])):
        #print(s, str(stocks[0][i]))
        if s == str(stocks[0][i]):
            return stocks[1][i]
# find_weight(stock02, 600000)

'''
merge 10 stocks in a year into one & assign weight accordingly then save it
'''

def merge(stocks, base_dir = '../components/',  year = 2002):
    year = str(year)
    folder = base_dir + year + "/"
    files = os.listdir(folder)
    df = pd.read_csv(os.path.join(folder, files[0]), index_col=0, parse_dates=True)
#     print(df)
    stock = files[0][:files[0].find('-')]
    weight = find_weight(stocks, stock)
    print (stock, weight)
    df['chg_amt'] = df['chg_amt'] * weight
    for i in range(1, len(files)):
        path = os.path.join(folder, files[i])
        # only parse csv file
        if path.find('csv') > 0:
            dfi = pd.read_csv(path, index_col=0, parse_dates=True)
            stock = files[i][:files[0].find('-')]
            weight = find_weight(stocks, stock)
            print (stock, weight)
            dfi['chg_amt'] = dfi['chg_amt'] * weight
            df = df.join(dfi, lsuffix='_l', rsuffix='_r')  #add suffix to differentiate data
    df = df.fillna(0)
    df.to_csv(folder + 'components'+'-'+year+'.csv')
    print('-------- Merge %s Done ------' % (year))

# merge all CSV in one folder into one
# merge(stock02, year=2002)
stocks = [stock03, stock04,stock05,stock06,stock07,stock08,stock09,stock10,stock11,
         stock12,stock13,stock14,stock15,stock16,stock17]
# for i in range(2003, 2018):
#     merge(stocks[i - 2003], year = i)
'''
Merge components in 10 year into one
'''    
def merge_all():
    base_dir = '../components/'
    file1 = '../components/2002/components-2002.csv'
    df = pd.read_csv(file1, index_col=0, parse_dates=True)
    for i in range(2003, 2018):
        year = str(i)
        file = base_dir + year + "/" + 'components'+'-'+year+'.csv'
        dfi = pd.read_csv(file, index_col=0, parse_dates=True)
        print('merge ', file)
        df = df.append(dfi)
    return df
# df = merge_all()
# print (df)
# df.to_csv('../components/components-top10.csv')

'''
Reading Top10 Components CSV
Data has been assigned weight according to their ratio in the market

# Ref: data calculated from data/generate/FetchingComponentsData.ipynb
'''
def read_components(base_dir = '../data/components/', 
                       filename = 'components-top10.csv',
                       start_date = '2002-01-04', end_date = '2017-11-30',
                       names = [i for i in range(10)]):
    filename = base_dir +'/'+ filename
    print (filename) 
    df = pd.read_csv(filename, index_col=0, 
                     skiprows=1,parse_dates=True,
                     names = names
                    )
    return df[start_date : end_date]
df = read_components('../components/',start_date='2005-01-04', end_date= '2015-12-31')
from sklearn import preprocessing
scaler = preprocessing.MinMaxScaler()
scaler.fit_transform(df)
```
